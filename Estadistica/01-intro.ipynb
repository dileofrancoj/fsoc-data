## Subplots
Con subplot() podes dibujar multiples gráficos en una sola figura.

La función subplot toma 3 argumentos posicionales que describen un **layout**. Este layout está organizando en filas y columnas que están representadas por el primer y el segundo argumento.

plt.subplot(1,2,1) la figura tiene 1 fila, 2 columnas y el gráfico es el primero (1,2,1)

## Tipos de gráficas

### Scatter

Es un tipo de gráfico que se usa para la representación de coordenadas cartesianas para mostrar valores entre 2 variables de un conjunto. Puede ser usado para variables continuas o discretas. Muy útil para entender correlaciones entre variables (regresión)

Los gráficos de dispersión, también denominados diagramas de dispersión o scatter plots en la literatura en inglés, se utilizan para comparar los valores que toman dos variables distintas, una de las cuales se representa a lo largo del eje x y la otra a lo largo del eje y. La gráfica resultante nos permite identificar visualmente la posible correlación entre las dos variables.

En mapplotlib, aun cuando es posible crear este tipo de gráficas con la ya conocida función plt.plot (basta ocultar las líneas y mostrar solo los puntos), existe una función mucho más potente: la función matplotlib.pyplot.scatter (funcionalidad también disponible como método de un conjunto de ejes). Esta función acepta como primeros dos argumentos los valores x e y de los puntos a mostrar. ### Bars

Un diagrama de barras es un gráfico usado para mostrar de forma resumida un grupo de datos que puede incluir variables cualitativas y cuantitativas. En general se usa para resumir una variable categórica con su frecuencia de aparición

### Histogramas

Se utiliza más para variables continuas. Es útil para ver la distribución que sigue una muestra.

Un histograma es una representación gráfica de la distribución de un conjunto de datos. Se utiliza comúnmente para visualizar la frecuencia con la que ocurren diferentes valores dentro de un conjunto de datos continuo. En un histograma, los datos se dividen en "bins" o intervalos y se cuenta cuántos valores caen en cada bin. Luego, se representan estas cuentas en un gráfico de barras.

Características clave de un histograma:

Eje x: En el eje horizontal, se colocan los valores de la variable que estás midiendo. Estos valores se dividen en intervalos o bins.

Eje y: En el eje vertical, se representa la frecuencia o el número de observaciones que caen en cada bin.

Barras: Cada bin se representa mediante una barra vertical. La altura de la barra corresponde a la frecuencia de observaciones en ese bin.

No hay espacios entre las barras: En un histograma, no hay espacios entre las barras, ya que los bins se tocan unos a otros.

Área bajo la curva: El área total bajo las barras del histograma es igual al número total de observaciones en el conjunto de datos.

Los histogramas son útiles para comprender la distribución de los datos y pueden proporcionar información sobre la tendencia central, la dispersión y la forma de la distribución. Por ejemplo, un histograma puede mostrar si los datos son simétricos, sesgados hacia la derecha o hacia la izquierda, o si siguen una distribución normal. También puedes identificar valores atípicos o picos en la distribución.

### Introducción a la estadística
La estadística es una ciencia que estudia la variabilidad, colección, organización, análisis, interpretación y presentación de los datos. También estudia el proceso aleatorio que los genera siguiendo leyes de probabilidad.

La estadística aplicada a las ciencias fácticas permite estudiar una determinada población a partir de la recopilación de información, el análisis de datos y la interpretación de resultados.

Es una ciencia esencial para el estudio cuantitativo de los fenómenos de masa o colectivos.

#### Áreas de la estadística
- Descriptiva
Se dedica a la descripción, visualización y resumen de datos orginados a partir de fenómenos de estudio.

- Inferencial
Se dedica a la generación de modelos y predicciones asociadas a los fenómenos en cuestión teniendo en cuenta la aleatoriedad de las observaciónes.

- Análisis exploratorio de datos
Se basa en la estadística descriptiva. El objetivo es, a través de un conjunto de datos, extraer características principales. Debemos estudiar la naturaleza de los datos antes de realizar supuestos y comenzar las hipótesis.

### Tipos de datos
- Cualitativos: Son las propiedades categóricas que pueden usarse para describir un ente (nacionalidad de una persona, tipo de fruta)

- Cuantitativa: Representan una cantidad. Presión arterial, pulsaciones, etc

#### Tipos de variables:
- Cualitativas nominales: Cuando los valores que toma esa variable permite clasificar a la población en categorias excluyentes. Ej: Grupo sanguineo, estado civil
- Cualitativas ordinales: Los valores pueden ser ordenados de acuerdo a un criterio. Ej: Paciente que ingresa a un hospital
- Cuantitativas discretas: Puede tomar solo algunos valores de un determinado intervalo de reales. Ej: Cantidad de pacientes que ingresaron al hospital, sueldo de personas sin decimales
- Cuantitativas continuas: Cuando pueden tomar todos los valores de un determinado intervalo de reales. Volumen, longitud, dinero, altura, etc

Poder identificar rápidamente el tipo de variable es importante ya que las librerias que utilizamos para la carga de datos depende mucho del tipo de dato. Por ejemplo, pandas optimiza mucho las operaciones cuando el tipo de dato es categórico
