```python
# quiero crear una función que me calcule el promedio
```

```python
def promedio(edades):
    totalEdades = 0
    cantidadEdades = len(edades)
    for edad in edades:
        totalEdades += edad # += acumulador
    print("El promedio de edades es: ", totalEdades/cantidadEdades)
```

```python
promedio(edades)
```

```python
for edad in edades:
    print("edad:", edad)
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,6,8,12]
y = [3,7,2,19,7]
```

```python
plt.plot(x,y)
plt.title("XvsY")
plt.xlabel("Cantidad de horas trabajadas")
plt.ylabel("Rendimiento laboral")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# pulsaciones (x) vs cansancio (y)
x = [80,86,91,100,104,120,160]
y = [50,65,67,80,81,89,95]
plt.plot(x,y, color='red',ms=10, marker='*')
plt.title("Cansancio vs Pulsaciones", loc="center") # titulo con location
plt.xlabel("Pulsaciones")
plt.ylabel("Cansancio")
plt.grid() # agregamos un grid
```

```python
# Gráfico de barras -> Cualitativa / Categórica
# frecuencia: cantidad de apariciones que tiene un elemento en un conjunto

categorias = ["Docencia","Gestion academica"]
cantidad = [3,10]

plt.bar(categorias,cantidad, color='skyblue')
plt.title("Categorias laborales FSOC")
plt.xlabel("Categorías")
plt.ylabel("Frecuencia")
plt.show()
```

### Tipo de datos
