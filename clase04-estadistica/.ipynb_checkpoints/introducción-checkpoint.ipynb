## Medidas de tendencia central

- Media: $\frac{\Sigma_{i=1} x_i}{N}$

Edades = {29,31,20,100}

Media / promedio / media aritmetica = (29 + 31 + 20, 100) / 4 = 45

Media: Es sensible a valores atípicos. No es robusta 


- Media recortada

Ordena un conjunto de datos

Edades = {20,29,31,32,35,54,60,61,72,100}

Recorta valores extremos: recortar un 10% de los valores extremos
Edades_recortada = {29,31,32,35,54,60,61,72}


- Mediana

La mediana representa el valor de posicion central en un conjunto de datos **ordenado**
No es sensible a valores atípicos

Edades = {7,7,10,12,13} -> impar

Mediana de edades = 10

Edades = {7,7,10,12,13,10000} -> par

Mediana Edades = (10 + 12) / 2 = 11

- Moda: Es el valor con mayor frecuencia de aparicion en un conjunto

Edad = {7,7,19}

La moda es 7

Edad = {7,7,19,19}

La moda es {7,19} -> bimodal

###  Reemplazar los datos nulos con algun criterio

Si la variable no tiene valores atípicos, vamos a reemplazar los faltantes por la **media**

Por otra parte, si la variable que estamos analizando cuenta con valores atípicos se reemplazaran los nulos por
- Mediana
- Media recortada
- IQR (Rango intercuartil)

```python
sns.boxplot(data=titanic, y='Age')
```

```python
print("la media del conjunto (edad)", titanic['Age'].mean())
print("la mediana del conjunto (edad)", titanic['Age'].median())
print("la moda del conjunto (edad)", titanic['Age'].mode())
```

```python
mediana_age = titanic['Age'].median()
titanic['Age'].fillna(mediana_age, inplace=True) # llenar los valores nulos o NaN
# inplace=True reemplaza sobre el mismo dataframe
# Para las funciones de reemplazo de valores nulos o NaN usemos el inplace=True
```
