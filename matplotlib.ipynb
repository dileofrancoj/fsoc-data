```python
# Cansancio vs pulsaciones
x = [62,80,86,95,98,102,105,120,160]
y = [20,50,67,74,76,85,88,92,95]
(len(x), len(y))
```

```python
import matplotlib.pyplot as plt # importar libreria
```

```python
plt.plot(x,y, color='red')
plt.title("Cansancio vs pulsaciones", loc="left")
plt.xlabel("Pulsaciones")
plt.ylabel("Cansancio")
plt.grid()
plt.show()
```

## Estadística

Es una rama de las matemáticas que permite recopilar, organiar, analizar datos

## Areas de la estadística

- Descriptiva: Se dedica a la descripcion, visualización y resumen de datos originados a partir de fenomenos de estudio

- Inferencial: Se dedica a la generacion de modelos y predicciones asociadas a fenomenos

## Tipos de datos
- Cualitativos / Categóricos: Son las propiedades categóricas que pueden usarse para describir un objeto (nacionalidad, sexo, tipo de fruta)
- Cuantitativos: Representan una cantidad (presion arterial, pulsaciones, edad peso)

## Tipos de variable

- Cualitativo nominal: Son propiedades categóricas que permiten clasificar a la poblacion en categorias mutuamente excluyente (grupo sanguineo) 
- Cualitativo ordinal: Los valores pueden ser orndeados de acuerdo a un criterio (encuesta, estado de un paciente)

- Cuantitativo discreto: Cantidad de pacientes que ingresa a un hospital, cantida de alumnos por curso, gente que vive en un lugar
- Cuantitativo continuo: Pueden tomar todos los valores de un determinado intervalo. Altura, tiempo, volumen, dinero
